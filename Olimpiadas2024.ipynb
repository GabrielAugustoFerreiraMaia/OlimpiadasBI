```python
import pandas as pd

df = pd.read_csv('/content/Olympics 2024.csv')

df.head()
```

```python

df.describe()
```

```python
# Medalhas totais por país
medalhas_totais = df.groupby('NOC')['Total'].sum().sort_values(ascending=False)
print(medalhas_totais)
```

```python
import matplotlib.pyplot as plt

# Gráfico das medalhas totais por país
plt.figure(figsize=(12, 8))
medalhas_totais.plot(kind='bar')
plt.title('Medalhas Totais por País')
plt.xlabel('País')
plt.ylabel('Total de Medalhas')
plt.xticks(rotation=90)
plt.show()
```

```python
# Medalhas totais por competição
medalhas_competicao = df.groupby('Competitions')['Total'].sum().sort_values(ascending=False)
print(medalhas_competicao)
```

```python
plt.figure(figsize=(12, 8))
medalhas_competicao.plot(kind='bar')
plt.title('Medalhas Totais por Competição')
plt.xlabel('Competição')
plt.ylabel('Total de Medalhas')
plt.xticks(rotation=90)
plt.show()
```

```python
# Medalhas por tipo
medalhas_por_tipo = df[['Gold', 'Silver', 'Bronze']].sum()
medalhas_por_tipo.plot(kind='bar')
plt.title('Distribuição das Medalhas por Tipo')
plt.xlabel('Tipo de Medalha')
plt.ylabel('Total de Medalhas')
plt.show()
```
